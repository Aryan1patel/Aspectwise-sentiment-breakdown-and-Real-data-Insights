Day 1 – Global EDA (Exploration)

Date:

Goal

Understand dataset contents and decide a suitable domain for ABSA.

Actions
	•	Streamed first 10,000 reviews for fast exploration
	•	Inspected available fields (reviewText, summary, overall, time, helpful)
	•	Counted unique ASINs to assess product diversity
	•	Used keyword heuristics to infer rough product categories

Results

Category distribution (10,000 reviews):
	•	Mobile Phones: 2,772
	•	Cameras: 1,791
	•	Cables & Adapters: 861
	•	Laptops: 623
	•	Headphones & Earbuds: 621
	•	GPS: 302
	•	Others / Mixed: 2,850

Key Observations
	•	Dataset contains multiple electronics categories
	•	Reviews vary widely in length and sentiment
	•	Star ratings are skewed positive and hide aspect-level dissatisfaction

Decision

Selected domain: Mobile Phones

Reason:
	•	Largest coherent category
	•	Rich multi-aspect opinions (battery, camera, screen, performance)
	•	High interview relevance and business value
